La derivada de f(x) en x=1, retorna un valor verdadero :

```
x=1
v_verdadero = -0.4*x**3-0.45*x**2-1*x-0.25
print(v_verdadero)
```

```
import numpy as np
import matplotlib.pyplot as plt

# INGRESO
xi=1
h=0.5
fx = lambda x: -0.1*x**4 - 0.15*x**3 - 0.5*x**2 - 0.25*x + 1.2

oh = (fx(xi+1*h)-fx(xi)) / h
oh2 = (-fx(xi+2*h) + 4*fx(xi+1*h) - 3*fx(xi))/2*h
print("oh: " + str(oh))
print("oh2: " + str(oh2))
```

**Error**


```
error =(v_verdadero-oh)
print("error con oh1: " + str(error))

error =(v_verdadero-oh2)
print("error con oh2: " + str(error))
```

```
import numpy as np
import matplotlib.pyplot as plt

# INGRESO
xi=1
h=0.5
fx = lambda x: -0.1*x**4 - 0.15*x**3 - 0.5*x**2 - 0.25*x + 1.2

oh = (fx(xi)-fx(xi-1*h)) / h
oh2 = (3*fx(xi) - 4*fx(xi-1*h) + fx(xi-2*h))/2*h
print("oh: " + str(oh))
print("oh2: " + str(oh2))
```

```
print("valor verdadero: " + str(v_verdadero))
error =(v_verdadero-oh)
print("error con oh1: " + str(error))

error =(v_verdadero-oh2)
print("error con oh2: " + str(error))
```
