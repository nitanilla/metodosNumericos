```
# ALGORITMO PROPIO ()

import numpy as np

def f(x):
  return np.sum(x**3 + 4*x**2 - 10)

def bisec(f,a,b,N=100):
  for k in range(N):
    x=(a+b)/2
    #derecho (canbio de signo)
    if f(x)*f(b)<0:
      a=x
    # izquierdo (cambio de signo)
    elif f(x)*f(a)<0:
      b=x
    else:
      break
    print(k,x,f(x))
bisec(f,1,2)


```

```
# ALGORITMO PROPIO () MAS GRAFICA 

import numpy as np
import matplotlib.pyplot as plt

def f(x):
  return np.sum(x**3 + 4*x**2 - 10)

def bisec(f,a,b,N=100):
  GX=[]
  GY=[]
  for k in range(N):
    x=(a+b)/2
    #derecho (canbio de signo)
    if f(x)*f(b)<0:
      a=x
    # izquierdo (cambio de signo)
    elif f(x)*f(a)<0:
      b=x
    else:
      break
    print(k,x,f(x))
    GX.append(x)
    GY.append(f(x))
  plt.plot(GX,GY)
  plt.axhline(0, color="black")

  plt.xlabel('x')
  plt.ylabel('y')
  plt.grid()
  plt.show()

bisec(f,1,2)
```

```
# Algoritmo de Bisección           ---------------------- CODIGO REUTILIZADO -----------------------
# [a,b] se escogen de la gráfica de la función
# error = tolera

import numpy as np
import matplotlib.pyplot as plt

# INGRESO
fx = lambda x: x**3 + 4*x**2 - 10 
a = 1
b = 2
tolera = 0.001

# PROCEDIMIENTO
tramo = b-a
while not(tramo<tolera):
    c = (a+b)/2
    fa = fx(a)
    fb = fx(b)
    fc = fx(c)
    cambia = np.sign(fa)*np.sign(fc)
    if cambia < 0: 
        a = a
        b = c
    if cambia > 0:
        a = c
        b = b
    tramo = b-a

# SALIDA
print('       raiz en: ', c)
print('error en tramo: ', tramo)
```

```
# Algoritmo de Bisección      ---------------------- CODIGO REUTILIZADO Y MEJORADO -----------------------
# [a,b] se escogen de la gráfica de la función
# error = tolera
import numpy as np

# INGRESO
fx = lambda x: x**3 + 4*x**2 - 10 
a = 1
b = 2
tolera = 0.001

# PROCEDIMIENTO
tabla = []
tramo = b-a

fa = fx(a)
fb = fx(b)
i = 1
while (tramo>tolera):
    c = (a+b)/2
    fc = fx(c)
    tabla.append([i,a,c,b,fa,fc,fb,tramo])
    i = i+1
                 
    cambia = np.sign(fa)*np.sign(fc)
    if (cambia<0):
        b = c
        fb = fc
    else:
        a=c
        fa = fc
    tramo = b-a
c = (a+b)/2
fc = fx(c)
tabla.append([i,a,c,b,fa,fc,fb,tramo])
tabla = np.array(tabla)

raiz = c

# SALIDA
np.set_printoptions(precision = 4)
print('[ i, a, c, b, f(a), f(c), f(b), tramo]')
# print(tabla)

# Tabla con formato
n=len(tabla)
for i in range(0,n,1):
    unafila = tabla[i]
    formato = '{:.0f}'+' '+(len(unafila)-1)*'{:.3f} '
    unafila = formato.format(*unafila)
    print(unafila)
    
print('raiz: ',raiz)
```

```
# GRAFICA   ---------------------- CODIGO REUTILIZADO -----------------------
import matplotlib.pyplot as plt

xi = tabla[:,2]
yi = tabla[:,5]

# ordena los puntos para la grafica
orden = np.argsort(xi)
xi = xi[orden]
yi = yi[orden]

plt.plot(xi,yi)
plt.plot(xi,yi,'o')
plt.axhline(0, color="black")

plt.xlabel('x')
plt.ylabel('y')
plt.title('Bisección en f(x)')
plt.grid()
plt.show()
```
